# Conclusion

Notre travail montre qu'il existe plusieurs méthodes permettant d'effectuer la tâche d'inférence en langage naturel (NLI). 
L'approche la moins performante est l'apprentissage en contexte. Ce résultat est attendu car c'est l'approche qui utilise directement le modèle pré-entraîné sans ajustement supplémentaire.
L'approche de fine-tuning de l'encodeur donne de bien meilleurs résultats, car elle permet au modèle de s'adapter spécifiquement à la tâche NLI avec une accuracy finale d'environ 76%. Le fine-tuning du modèle complet offre des bonne performances avec une accuracy d'environ 70%.
